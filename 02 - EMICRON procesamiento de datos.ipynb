<div style="text-align: center; color: green; font-size: 25px; line-height: 1.5;">Crear columna con los nombres en los sectores basado en los codigos CIIU</div>

***La siguiente pieza de codigo crea una nueva columna llamada nombre_grupo  en el dataset df_vi_mi , esta nueva columna contiene los nombres de los grupos correspondientes a cada codigo en la columna original GRUPOS 12, ejemplo: si en la columna GRUPOS12 esta el valor 1, la nueva columna que creamos  nombre_grupo contiene el nombre correspondiente a dicho codigo :  'Agricultura, ganadería, caza, silvicultura y pesca'***

```python
def map_grupos12(value):
    mapping = {
        1: 'Agricultura, ganadería, caza, silvicultura y pesca',
        2: 'Minería',
        3: 'Industria manufacturera',
        4: 'Construcción',
        5: 'Comercio y reparación de vehículos automotores y motocicletas',
        6: 'Transporte y almacenamiento',
        7: 'Alojamiento y servicios de comida',
        8: 'Información y comunicaciones',
        9: 'Actividades inmobiliarias, profesionales y servicios administrativos',
        10: 'Educación',
        11: 'Actividades de atención a la salud humana y de asistencia social',
        12: 'Actividades artísticas, de entretenimiento, de recreación y otras actividades de servicios'
    }
    return mapping.get(value, 'Unknown')

df_vi_mi['nombre_grupo'] = df_vi_mi['GRUPOS12'].apply(map_grupos12, meta=('GRUPOS12', 'object'))
```

    Una vez hemos cargado y unido los datasets con dask, procederemos a realizar calculos, agregaciones, conteos, promedios y demas calculos descriptivos sobre el conjunto de datos. Es importante resaltar que todos estos calculos agregaciones y procedimientos podriamos hacerlos con dask, sin embargo es un tanto complejo y verboso el codigo para realizar los calculos lo cual dificulta en terminos practicos el analisis. Por fortuna podemos usar SQL  dentro de los datos que hemos cargado en dask, el cual es un lenguaje de procesamiento de datos mucho mas simple para hacer todos estos mismos calculos, agregaciones y demas operaciones sobre los datos de una forma mucho mas efectiva y practica.
Tanto con dask como con sql podemos hacer los mismos  calculos , sin embargo es mucho mas sencillo hacerlos con sql
</div>


```python
# Iniciar contexto SQL
context = Context()

# Registrar nuestra tabla para poder consultarla en SQL, cuando tenemos una tabla de dask que queremos consultar
# con SQL la registramos con el comando context.create_table() añadiendo el nombre de los datoos.
context.create_table("df_vi_mi", df_vi_mi)
```

## Calculo de valor agregado total  de micronegocios por sectores y participacion porcentual (GRUPOS12)

```python
# Ejecutar la sentencia SQL
result = context.sql(
    '''SELECT
        SUM("VALOR_AGREGADO_ANUAL_FEX") as valor_agregado,
        "nombre_grupo"
        FROM df_vi_mi a
        GROUP BY "nombre_grupo"
        order by 2''')
result.compute()
```

```python
#De la misma forma , podemos guardar el resultado de una consulta anterior
context.create_table("result", result)

valagre_sectores = context.sql(
    '''SELECT
            ("valor_agregado"/77221773573221.469)*100 as porcentaje_val_agregado,
            "nombre_grupo"
            FROM result order by 1 desc
    '''
)


valagre_sectores.compute()
```

## Calculo de cantidad total de micronegocios por sectores y participacion porcentual (GRUPOS12)

```python
# Ejecutar la sentencia SQL
cant_empresas = context.sql(
    '''SELECT
        SUM("F_EXP_x") as cantidad_empresas,
        "nombre_grupo"
        FROM df_vi_mi a
        GROUP BY "nombre_grupo"
        order by 2''')
cant_empresas.compute()
```

```python
#De la misma forma , podemos guardar el resultado de una consulta anterior
context.create_table("cant_empresas", cant_empresas)

cantidad_empresas_sectores = context.sql(
    '''SELECT
            ("cantidad_empresas"/5188402.435524806)*100 as porcentaje_cant_empresas,
            "nombre_grupo"
            FROM cant_empresas order by 1 desc
    '''
)


cantidad_empresas_sectores.compute()
```

# Cantidad de empresas segun situacion en el empleo del propietario en cabeceras municipales

```python
# Ejecutar la sentencia SQL
cant_empresas_cabecera_situacion_propietario = context.sql(
    '''SELECT
        SUM("F_EXP_x") as cantidad_empresas,
        "nombre_P3033"
        FROM df_vi_mi a
        WHERE "CLASE_TE_x"=1
        GROUP BY 2
        order by 2''')
cant_empresas_cabecera_situacion_propietario .compute()
```

```python
context.create_table("cant_empresas_cabecera_situacion_propietario", cant_empresas_cabecera_situacion_propietario)

porcentaje_empresas_situacion_cabecera = context.sql(
    '''SELECT
            round("cantidad_empresas"/3659977.667*100,1) as porcentaje_cabecera,
            "nombre_P3033"
            FROM cant_empresas_cabecera_situacion_propietario order by 1 desc
    '''
)
porcentaje_empresas_situacion_cabecera.compute()
```

# Cantidad de empresas segun situacion en el empleo del propietario en RURAL DISPERSO

```python
# Ejecutar la sentencia SQL
cant_empresas_rural_situacion_propietario = context.sql(
    '''SELECT
        SUM("F_EXP_x") as cantidad_empresas,
        "nombre_P3033"
        FROM df_vi_mi a
        WHERE "CLASE_TE_x"=2
        GROUP BY 2
        order by 2''')
cant_empresas_rural_situacion_propietario.compute()
```

```python
context.create_table("cant_empresas_rural_situacion_propietario", cant_empresas_rural_situacion_propietario)

porcentaje_empresas_situacion_rural = context.sql(
    '''SELECT
            round("cantidad_empresas"/1528424.7689999999*100,1) as porcentaje_rural,
             "nombre_P3033"
            FROM cant_empresas_rural_situacion_propietario order by 1 desc
    '''
)
porcentaje_empresas_situacion_rural.compute()
```
